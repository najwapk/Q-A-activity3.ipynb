```python
import pandas as pd
diab=pd.read_csv('C:\\Users\\91735\\OneDrive\\Desktop\\diabetes_model.csv')
diab
```

```python
# 1)Display the Glucose level and the Age for the samples having the index ranging from 45 to 60.
diab[['Glucose','Age']][45:61]
```

```python
# 2)How many patients in the data are having the blood pressure level above 120?

no = 0
for bp in diab.BloodPressure:
    if bp>120:
        no+=1
print('Number of Patients ',no) 
diab[(diab['BloodPressure']>120)]
```

```python
# 3) Select the rows the age is between 24 and 29 (inclusive)

diab[(diab['Age']>=24) & (diab['Age']<=29)]
```

```python
# 4) Find the minimum and maximum value of insulin level in the data.

print('Minimum value of insulin is',diab['Insulin'].min())
```

```python
print('Maximum value of insulin is',diab['Insulin'].max())
```

```python
# 5) Create a histogram of "Age".

import matplotlib.pyplot as plt
plt.figure(figsize =(6,6))
plt.hist(diab['Age'],rwidth=0.7)
plt.title('Age Distribution of Diabetics Patients ')
plt.xlabel('Age')
plt.ylabel('Frequency')
```

```python
# 6) Create a scatter plot of "BMI" vs "Glucose".

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(6,6))
plt.scatter(diab['BMI'],diab['Glucose'],s=10,c='red')
plt.title('Scatter plot of BMI vs Glucose')
plt.xlabel('BMI')
plt.ylabel('Glucose')
```

```python
#7) Replace the min value of "BMI" with mean value of BMI

print('The current min value of BMI is',diab['BMI'].min())
diab['BMI'].mean()
diab['BMI'].replace(diab['BMI'].min(),diab['BMI'].mean(),inplace=True)
print('The new value is',diab['BMI'].min())
```

```python
# 8) Draw a grouped boxplot with "Outcome" on x-axis and "Glucose" on the y-axis such that we see distributions of Glucose separately for each outcome.

import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(data=diab,x='Outcome',y='Glucose')
```
